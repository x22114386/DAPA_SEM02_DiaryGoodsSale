```python
# Data Type Conversion
dairy_df['Date'] = pd.to_datetime(dairy_df['Date'])  # Convert 'Date' column to datetime

# Convert 'date' column to datetime
dairy_df['Production Date'] = pd.to_datetime(dairy_df['Production Date'])

# Removing Outliers (example using z-score)
z_scores = (dairy_df['Quantity (liters/kg)'] - dairy_df['Quantity (liters/kg)'].mean()) / dairy_df['Quantity (liters/kg)'].std()
dairy_df = dairy_df[z_scores.abs() < 3]  # Remove rows with z-scores outside the threshold

# Standardizing and Normalizing Data
dairy_df['Quantity (liters/kg)'] = (dairy_df['Quantity (liters/kg)'] - dairy_df['Quantity (liters/kg)'].mean()) / dairy_df['Quantity (liters/kg)'].std()

# Handling Inconsistent Values
dairy_df['Location'] = dairy_df['Location'].str.strip()  # Remove leading/trailing white spaces

# Data Formatting
dairy_df['Product Name'] = dairy_df['Product Name'].str.title()  # Capitalize each word in 'Product Name'

# Handling Encoding Issues (example using UTF-8 encoding)
dairy_df['Product Name'] = dairy_df['Product Name'].str.encode('utf-8').str.decode('utf-8')

# Feature Engineering
dairy_df['Total Revenue (INR)'] = dairy_df['Quantity (liters/kg)'] * dairy_df['Price per Unit']

# Data Validation (example checking minimum stock threshold)
dairy_df = dairy_df[dairy_df['Quantity in Stock (liters/kg)'] >= dairy_df['Minimum Stock Threshold (liters/kg)']]

# Display the cleaned dataset
print(dairy_df.head())
```

```python
# Removing Outliers using z-score
z_scores = (dairy_df['Quantity (liters/kg)'] - dairy_df['Quantity (liters/kg)'].mean()) / dairy_df['Quantity (liters/kg)'].std()
outliers = dairy_df[z_scores.abs() >= 3]  # Identify rows with z-scores outside the threshold

# Plotting the outliers
plt.scatter(dairy_df.index, dairy_df['Quantity (liters/kg)'], label='Data')
plt.scatter(outliers.index, outliers['Quantity (liters/kg)'], color='red', label='Outliers')
plt.xlabel('Index')
plt.ylabel('Quantity (liters/kg)')
plt.title('Outliers')
plt.legend()
plt.show()
```

```python
# Specify the maximum number of categories to show individually
max_categories = 5

cat_vars = ['Location', 'Farm Size', 'Product Name',
            'Brand', 'Storage Condition', 'Customer Location',
            'Sales Channel']

# Create a figure and axes
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

# Create a pie chart for each categorical variable
for i, var in enumerate(cat_vars):
    if i < len(axs.flat):
        # Count the number of occurrences for each category
        cat_counts = dairy_df[var].value_counts()

        # Group categories beyond the top max_categories as 'Other'
        if len(cat_counts) > max_categories:
            cat_counts_top = cat_counts[:max_categories]
            cat_counts_other = pd.Series(cat_counts[max_categories:].sum(), index=['Other'])
            cat_counts = pd.concat([cat_counts_top, cat_counts_other])

        # Create a pie chart
        axs.flat[i].pie(cat_counts, labels=cat_counts.index, autopct='%1.1f%%', startangle=90)

        # Set a title for each subplot
        axs.flat[i].set_title(f'{var} Distribution')

# Adjust spacing between subplots
fig.tight_layout()

# Remove eighth plot
fig.delaxes(axs[2][1])

# Remove ninth plot
fig.delaxes(axs[2][2])

# Show the plot
plt.show()
```

# Data split as Train and Test

```python
# Remove Unnecesary column / attribute
dairy_df.drop(columns=['Product ID', 'Production Date'], inplace=True)
# Remove Date and Expiration Date, but keep Production Date for Exploratory Data Analysis
dairy_df.drop(columns=['Date', 'Expiration Date'], inplace=True)
dairy_df.head()
```

```python
from sklearn.model_selection import train_test_split

# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(dairy_df.drop('Reorder Quantity (liters/kg)', axis=1), dairy_df['Reorder Quantity (liters/kg)'], test_size=0.2, random_state=0)
# Concatenate X_train and y_train for outlier removal
train_df = pd.concat([X_train, y_train], axis=1)

# Calculate the IQR values for each column
Q1 = train_df.quantile(0.25)
Q3 = train_df.quantile(0.75)
IQR = Q3 - Q1

# Remove outliers from X_train
train_df = train_df[~((train_df < (Q1 - 1.5 * IQR)) | (train_df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Separate X_train and y_train after outlier removal
X_train = train_df.drop('Reorder Quantity (liters/kg)', axis=1)
y_train = train_df['Reorder Quantity (liters/kg)']
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import RidgeCV
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Create base regressor models
estimators = [
    ('ridge', make_pipeline(StandardScaler(), RidgeCV())),
    ('svr', make_pipeline(StandardScaler(), SVR())),
    ('rf', RandomForestRegressor())
]

# Create StackingRegressor model
stacking_regressor = StackingRegressor(estimators=estimators, final_estimator=RandomForestRegressor())

# Fit the StackingRegressor to the training data
stacking_regressor.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred_stacking = stacking_regressor.predict(X_test_scaled)

# Calculate evaluation metrics for StackingRegressor
mae_stacking = metrics.mean_absolute_error(y_test, y_pred_stacking)
mape_stacking = mean_absolute_percentage_error(y_test, y_pred_stacking)
mse_stacking = metrics.mean_squared_error(y_test, y_pred_stacking)
r2_stacking = metrics.r2_score(y_test, y_pred_stacking)
rmse_stacking = math.sqrt(mse_stacking)

print('Stacking Regressor - MAE:', mae_stacking)
print('Stacking Regressor - MAPE:', mape_stacking)
print('Stacking Regressor - MSE:', mse_stacking)
print('Stacking Regressor - R2:', r2_stacking)
print('Stacking Regressor - RMSE:', rmse_stacking)
```
